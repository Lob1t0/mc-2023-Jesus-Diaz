```python
import math
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6, 7]
y = [0.5, 2.5, 2, 4, 3.5, 6, 5.5]
n = len(x)
sum_x = sum(x)
sum_y = sum(y)
sum_xy = sum([xi * yi for xi, yi in zip(x, y)])
sum_xx = sum([xi ** 2 for xi in x])
mean_x = sum_x / n
mean_y = sum_y / n
a1 = (n * sum_xy - sum_x * sum_y) / (n * sum_xx - sum_x ** 2)
a0 = mean_y - a1 * mean_x
regression_line = [a0 + a1 * xi for xi in x]
y_mean = sum_y / n
St = sum([(yi - y_mean) ** 2 for yi in y])
Sy = math.sqrt(St / (n - 1))
Sr = sum([(yi - a0 - a1 * xi) ** 2 for xi, yi in zip(x, y)])
Syx = math.sqrt(Sr / (n - 2))
r = math.sqrt((St - Sr) / St) * 100
print(f"Regression line: y = {a0:.4f} + {a1:.4f}x")
print(f"Standard deviation (Sy): {Sy:.4f}")
print(f"Standard error of the estimate (Sy/x): {Syx:.4f}")
print(f"Coefficient of correlation (r): {r:.4f}%")
plt.scatter(x, y, color='black')
plt.plot(x, regression_line, color='blue')
plt.title("Linear Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
