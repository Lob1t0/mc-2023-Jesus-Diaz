##Punto dos

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([1.8, 2, 2.2, 2.6, 2.9, 3.3, 3.7, 4.1])


linear_coeffs = np.polyfit(x, y, 1)
linear_func = np.poly1d(linear_coeffs)
linear_mse = np.mean((linear_func(x) - y) ** 2)


quadratic_coeffs = np.polyfit(x, y, 2)
quadratic_func = np.poly1d(quadratic_coeffs)
quadratic_mse = np.mean((quadratic_func(x) - y) ** 2)


x_vals = np.linspace(1, 8, 100)


plt.plot(x, y, 'o', label='Puntos dados')
plt.plot(x_vals, linear_func(x_vals), label='Función lineal')
plt.plot(x_vals, quadratic_func(x_vals), label='Polinomio de grado 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

print("Error cuadrático medio de la función lineal:", linear_mse)
print("Error cuadrático medio del polinomio de grado 2:", quadratic_mse)


```

##Punto tres

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline


x = np.array([1, 2, 3, 4, 5])
y = np.array([1.8, 2, 2.5, 2.8, 3.5])


cs = CubicSpline(x, y)


x_val = 3.2
y_val = cs([x_val])
print("f(3.2) =", y_val)


x_vals = np.linspace(1, 5, 100)
y_vals = cs(x_vals)

plt.plot(x, y, 'o', label='Puntos dados')
plt.plot(x_vals, y_vals, label='Trazadores cúbicos')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()


```
